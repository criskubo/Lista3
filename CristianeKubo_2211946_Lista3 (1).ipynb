```python
#1. Salve a base de dados escolhida no seu drive e importe para um notebook do Colab, criando um Dataframe
import pandas as pd
from google.colab import drive
drive.mount('/drive')
arq=pd.read_csv("/drive/MyDrive/class_analise_credito.csv")
display(arq)
```

```python
#2. Inspecione as colunas do conjunto de dados (Nome das colunas e tipo de dados)
dados=arq.info()
print(dados)
```

```python
#3. Escolha uma coluna com valores numéricos e apague todas as linhas que possuam valor nulo
novaCol=arq['RENDA'].dropna()
display(arq['RENDA'])
```

```python
#4. Calcule a média desta mesma coluna
media=arq['RENDA'].mean()
print(media)
```

```python
#5. Escolha um critério para eliminar as linhas que possuam valores muito distantes da média e crie um novo Dataframe.
vDist1=arq[arq['RENDA']<=media+500].index
vDist2=arq[arq['RENDA']>=media-500].index
arq['RENDA'].drop(vDist1,inplace=True)
arq['RENDA'].drop(vDist2,inplace=True)
novoDf=pd.DataFrame(arq)
print(novoDf)
```

```python
#6. Crie um novo Dataframe que contenha apenas 5 colunas do conjunto anterior.
col1=novoDf['NDEP']
col2=novoDf['RENDA']
col3=novoDf['TIPOR']
col4=novoDf['VBEM']
col5=novoDf['NPARC']
segnovoDf=pd.concat([col1,col2,col3,col4,col5],axis=1)
pd.set_option('display.max_columns',0)
pd.set_option('display.max_rows',5)
print(segnovoDf)
```

```python
#7. Escolha duas colunas numéricas e crie uma nova coluna que siga a regra: (Coluna A - Coluna B)*número aleatório (entre 0 e 1).
segnovoDf['colNova']=(col1-col2)*(0.5)
display(segnovoDf)
```

```python
#8. Embaralhe aleatoriamente o dataframe (índices e valores).
import numpy as np

dfEmb=segnovoDf.reindex(np.random.permutation(segnovoDf.index))
display(dfEmb)
```

```python
#9. Ordene o Dataframe seguindo um critério definido por você (Explique o critério)
#O critério escolhido organiza o Dataframe em ordem decrescente dos valores que correspondem à renda.
ordena_arq=segnovoDf.sort_values(by='RENDA',ascending=False)
pd.set_option('display.max_columns',6)
pd.set_option('display.max_rows',5)
display(ordena_arq)
```

```python
#11. Utilize a função MAPS para fazer algum tipo de substituição no Dataframe (explique)
#A função MAPS foi utilizada para substituir os valores correspondentes a 9 da coluna 'NPARC' para 10.

segnovoDf['NPARC']=segnovoDf['NPARC'].map({9:10})
pd.set_option('display.max_columns',6)
pd.set_option('display.max_rows',10)
display(segnovoDf)
```
